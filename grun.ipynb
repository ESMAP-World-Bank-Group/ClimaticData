```python
import xarray as xr
import geopandas as gpd
from utils import *
```

```python
grun_file = 'GRUN/GRUN_v1_GSWP3_WGS84_05_1902_2014.nc'

ISO_A2 = 'GN'
long_west, lat_south, long_east, lat_north = get_bbox(ISO_A2)
print(long_west, lat_south, long_east, lat_north)

catchment_geojson = 'catchment_files/stationbasins_guinea.geojson'
data_catchment = gpd.read_file(catchment_geojson)

folder_output = 'GRUN/output'
if not os.path.exists(folder_output): os.makedirs(folder_output)
```

```python
data_grun = xr.open_dataset(grun_file)

calculate_resolution_netcdf(data_grun)

# Subset the dataset
data_grun = data_grun.sel(
    X=slice(long_west, long_east),
    Y=slice(lat_south, lat_north)  # Note: Y may be decreasing
)
```

```python
data_grun
```

```python
plot_spatial_mean_timeseries_all_vars(data_grun, lat_name='Y', long_name='X', folder=folder_output)
plot_monthly_climatology_grid(data_grun, 'Runoff', folder=folder_output)
```

```python
df = calculate_spatial_mean_annual(data_grun, data_catchment, lat_name='Y', long_name='X')
df.to_csv(os.path.join('GRUN', f'GRUN_{ISO_A2}_runoff_monthly.csv'), index=False)

df_yearly = convert_monthly_mm_day_to_yearly_mm_year(df.copy(), var_name="Runoff")
df_yearly.replace(0, pd.NA, inplace=True)
df_yearly.pivot(index='year', columns='region').to_csv(os.path.join(folder_output, f'GRUN_{ISO_A2}_runoff_mm-year.csv'), index=True)
```
